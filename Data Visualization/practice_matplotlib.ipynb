```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Basic

### Task 1: Line Plot

    CSV: sales_data.csv

    Goal: Plot sales across months.

    Hints: plt.plot(), marker='o', add grid.

```python
df = pd.read_csv('Practice_Datasets/sales_data.csv')
print(df)
```

```python
plt.figure(figsize=(12,9))
plt.plot(df['Month'],df['Sales'] ,marker='o')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.title('Sales Across months.')
plt.show()
```

### Task 2: Bar Chart

    CSV: sales_data.csv

    Goal: Show sales in a bar chart (Month vs Sales).

```python
plt.figure(figsize=(12,10))
plt.bar(df['Month'],df['Sales'], edgecolor= 'black')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(axis='y')
plt.title("Show Sales data per Month")
plt.xticks(rotation = 45)
plt.show()
```

### Task 3: Horizontal Bar Chart

    CSV: students_scores.csv

    Goal: Plot students' Maths scores as horizontal bars.

```python
stud = pd.read_csv('Practice_Datasets/students_scores.csv')

plt.figure(figsize=(12,5))
plt.barh(stud['Student'],stud['Maths'],edgecolor='black')
plt.xlabel('Maths Marks')
plt.ylabel('Students')
plt.title('All Students Maths Marks')
plt.grid(axis='x')
plt.show()
```

### Task 4: Pie Chart

    CSV: sales_data.csv

    Goal: Show percentage contribution of each month's sales.


```python
plt.pie(df['Sales'],labels=df['Month'],autopct='% 1.1f%%',startangle=90)
plt.title('Percentage Contribution of Sales in Each Month')
plt.show()
```

### Task 5: Scatter Plot

    CSV: students_scores.csv

    Goal: Plot Maths vs Physics scores for students.

    Hint: plt.scatter(x, y)
|

```python
plt.scatter(stud['Maths'],stud['Physics'],edgecolors='black')
plt.xlabel('Maths Scores')
plt.ylabel('Physics Scores')
plt.title('Maths vs Physics Scores')
plt.grid(True)
plt.show()
```

### Task 6: Histogram

    CSV: students_scores.csv

    Goal: Plot a histogram of Chemistry scores.

```python
plt.figure(figsize=(6,5))
plt.hist(stud['Chemistry'],bins=5, edgecolor='black')
plt.xlabel('Chemistry Scores')
plt.ylabel('Number of Students')
plt.title('Distribution of Chemistry Scores')
plt.grid(axis='y')
plt.show()
```

# Intermediate

### Task 7: Subplots

    CSV: sales_data.csv

    Goal:

        First plot: Line chart

        Second plot: Bar chart

    Hint: plt.subplot(1, 2, 1)

```python
fig,axes = plt.subplots(1,2,figsize=(12,6))
axes[0].plot(df['Month'],df['Sales'],marker='o')
axes[0].set_xlabel('Months')
axes[0].set_ylabel('Sales Data')
axes[0].set_title('Sales Per Month')
axes[0].grid(axis='y')
axes[0].set_xticklabels(df['Month'], rotation=90)

axes[1].bar(df['Month'],df['Sales'])
axes[1].set_xlabel('Months')
axes[1].set_ylabel('Sales Data')
axes[1].set_title('Sales Per Month')
axes[1].grid(axis='y')
axes[1].set_xticklabels(df['Month'],rotation=90)

```

### Task 8: Stacked Bar Chart

    CSV: students_scores.csv

    Goal: Plot stacked bars of Maths, Physics, and Chemistry scores for each student.


```python
plt.figure(figsize=(10,8))
plt.bar(stud['Student'],stud['Maths'], label='Maths')
plt.bar(stud['Student'],stud['Physics'],bottom=stud['Maths'],color='Orange',label='Physics')
plt.bar(stud['Student'],stud['Chemistry'],bottom= stud['Maths'] +stud['Physics'],color='Purple',label='Chemistry')

plt.xlabel("Student")
plt.ylabel("Marks")
plt.title('Stack Score of Three subject Maths, Physics and Chemistry')
plt.legend()
plt.grid(axis='y')
plt.show()
```

### Task 9: Customize Plot

    CSV: sales_data.csv

    Goal:

        Customize Line Plot

        Add title, labels, legend, change line style, color.

```python
plt.figure(figsize=(10,8))
plt.plot(df['Month'],df['Sales'],linestyle='--', color='green', marker='o', label='Sales Over Months')
plt.xlabel('Months', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.title('Months and Sales Data', color='red', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

### Task 10: Box Plot

    CSV: students_scores.csv

    Goal: Create a boxplot showing score distribution for Maths, Physics, Chemistry.

```python
plt.figure(figsize=(10,8))
plt.boxplot([stud['Maths'], stud['Physics'], stud['Chemistry']], labels=['Maths', 'Physics', 'Chemistry'])

plt.title('Score Distribution in Maths, Physics, and Chemistry', fontsize=16)
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

### Task 11: Heatmap

    CSV: temperature_data.csv

    Goal: Create a heatmap showing temperature changes city-wise.

    Hint: Use plt.imshow() for a simple heatmap.

```python
temp = pd.read_csv('Practice_Datasets/temperature_data.csv')
print(temp)
```

```python
correlation = temp[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']].corr()
plt.imshow(correlation,cmap='coolwarm',interpolation=None)
plt.xticks(range(len(correlation)),correlation.columns,rotation = 90)
plt.yticks(range(len(correlation)),correlation.columns)
plt.colorbar()
plt.title('HeatMap of Temperature..')
plt.show()
```
