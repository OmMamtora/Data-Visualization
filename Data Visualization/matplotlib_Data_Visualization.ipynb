```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/diamonds.csv') 
print(df.head())
```

```python
# Line Plot

plt.plot(df['carat'],df['price'])
plt.title('Carat VS Price Line Plot')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter Plot

plt.scatter(df['carat'],df['price'])
plt.title('Carat VS Price Scatter Plot')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# bar Plot

# Here we are takign cut on x axis and we have to find average price of cut so we are using groupby method for
# finding average price of cut.


plt.bar(df['cut'].unique(),df.groupby('cut')['price'].mean())
plt.title('Average price by Cut')
plt.xlabel('Cut')
plt.ylabel('Average price')
plt.show()
```

### Intermediate Plots using Matplotlib

```python
# Hitogram

# Bins: Bins are the ranges that group data points in a histogram to show how frequently values occur within each
# range.


plt.hist(df['price'],bins=10,edgecolor= "black")
plt.title('Distribution of prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Histogram

plt.hist(df['price'],bins=200,edgecolor= "black")
plt.title('Distribution of prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot

plt.boxplot(df['carat'],vert=False)  # Return Horizontal box
plt.title('Box plot of Carat')
plt.xlabel('Carat')
plt.show()
```

```python

plt.boxplot(df['carat'])  # Return Vertical Box
plt.title('Box plot of Carat')
plt.xlabel('Carat')
plt.show()
```

### Advance Plots

```python
# HeatMap

correlation = df[['carat','depth','table','price','x','y','z']].corr()
print(correlation)
```

----------------

cmap='coolwarm' used for color cool-> blue color and warm -> red color

correlation.columns :-> used for give column name because all column has their name

rotation = 90 :-> this is usd for rotating text on Map

#### Here, Dark red color means strong positive correlation  close to +1.

#### Light colors(white/light pink/blue)	:-> Weak or no correlation  Close to 0.

#### Dark Blue menas Strong negative correlation (as one value increases, the other decreases)  Close to -1	.


```python
correlation = df[['carat','depth','table','price','x','y','z']].corr()
plt.imshow(correlation,cmap='coolwarm',interpolation='None')
plt.colorbar()
plt.xticks(range(len(correlation)),correlation.columns,rotation = 90)
plt.yticks(range(len(correlation)),correlation.columns)
plt.title('HeatMap of Correlation Matrix..')
plt.show()
```

```python
# Sub Plot

fig, axes = plt.subplots(1,2,figsize = (15,5))
axes[0].hist(df['carat'],bins=20,edgecolor = 'black')
axes[0].set_title("Histogram of carat")
axes[0].set_xlabel('carat')
axes[1].hist(df['price'],bins=10,edgecolor='black')
axes[1].set_xlabel('Price')
axes[1].set_title('Histogram of Price')
plt.show()
```

```python
# 2 X 2 Subplot

fig,axes = plt.subplots(2,2,figsize=(12,10))
axes[0,0].hist(df['carat'],bins=10,edgecolor='black')
axes[0,0].set_xlabel(' 0 , 0')
axes[0,0].set_title('Histogram of carat')

axes[0,1].hist(df['price'],bins=20,edgecolor= 'black')
axes[0,1].set_xlabel(' 0 , 1')
axes[0,1].set_title('Histogram of Price')


axes[1,0].scatter(df['carat'],df['price'])
axes[1,0].set_xlabel('Carat')
axes[1,0].set_ylabel('Price')
axes[1,0].set_title('Carat and price Scatter Plot')

axes[1,1].bar(df['cut'].unique(),df.groupby('cut')['price'].mean())
axes[1,1].set_title('Average Price by Cut')
axes[1,1].set_xlabel('cut')
axes[1,1].set_ylabel('Average Price per Cut')
```
