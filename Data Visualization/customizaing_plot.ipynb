# Customizing Plots in Seaborn And Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/diamonds.csv')
print(f"Number of rows before sampling : {df.shape[0]}")
print(f"Number of columns before sampling : {df.shape[1]}")
print(f"{50*'-'}")
df = df.sample(5000)
print(f"Number of rows after sampling : {df.shape[0]}")
print(f"Number of columns after sampling : {df.shape[1]}")

print(df.head())
```

### 1. Setting Title

```python
sns.scatterplot(data=df,x='carat',y='price')
plt.title('Scatter plot of Carat VS Price')
plt.show()
```

###  2.Settings Axes Labls


```python
sns.scatterplot(data=df,x='carat',y='price')
plt.xlabel('Carat Weight')
plt.ylabel('Price(in USD)')
plt.show()
```

### 3. Customizing the Legend

```python
sns.scatterplot(data=df,x='carat',y='price',hue='cut')
plt.legend(title='Diamond Cut',loc='upper right')
plt.show()
```

### 4. Customizing Ticks

```python
sns.histplot(df['price'],bins=10,edgecolor='black')
plt.xlabel('Price',fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

### 5. Adding a Grid

```python
sns.set_style('whitegrid')
sns.scatterplot(data=df,x='carat',y='price')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

### 6. Customizing Spines

```python
sns.boxplot(data=df,x='cut',y='price')
sns.despine()
plt.show()
```

```python
sns.boxplot(data=df,x='cut',y='price',palette='Set3',hue='cut')
sns.despine()  # Remove the top and right spine for cleaner look
plt.show()
```

### 7. Setting Figure Size

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='cut',y='price')
plt.title('Bar Plot of price by Cut (Large Figure)')
plt.show()
```

### 8. Creating SubPlotas

```python
fig,axes = plt.subplots(1,2,figsize=(12,5))
sns.histplot(df['carat'],bins=10,edgecolor='black',ax=axes[0],kde=True)
axes[0].set_title('Histogram of Carat')

sns.histplot(df['price'],bins=10,edgecolor='black',ax=axes[1],kde=True)
axes[1].set_title('Histogram of Price')

plt.show()
```
