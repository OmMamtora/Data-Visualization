```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Seaborn

### Basics

```python
df = pd.read_csv('Datasets/diamonds.csv')
print(df.head())
```

```python
# Line Plot

sns.lineplot(x='carat', y='price', data=df)
plt.title('Seaborn Line Plot of carat and price')
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(data=df, x='carat',y='price')
plt.title('Seaborn Scatter Plot of Carat VS Price')
plt.show()
```

### hue: used to visualize the data of different categories in one plot

```python
sns.scatterplot(data=df, x='carat',y='price',hue='cut')
plt.title('Seaborn Scatter Plot of Carat VS Price')
plt.show()
```

```python
# Seaborn Line Plot with Facets basedd on 'cut' column

g = sns.FacetGrid(df,col='cut', col_wrap=2, height=4)
g.map(sns.lineplot,'carat','price')
g.add_legend()
plt.show()
```

### Intermediate

```python
# Only One Box Plot
sns.boxplot(x='price',data=df)
plt.title("Box Plot of Price")
plt.show()
```

```python
# Box Plot

sns.boxplot(data=df,x='cut',y='price')
plt.title('Box Plot of Cut and Price')
plt.show()
```

```python
# Violin Plot

sns.violinplot(data=df, x='color',y='price')
plt.title('Violin plot of price by color')
plt.show()
```

```python
# Count Plot

sns.countplot(data=df,x='cut')
plt.title('Count plot of cut')
plt.show()
```

```python
sns.countplot(data=df,x='cut',hue='color')
plt.title('Count plot of cut')
plt.show()
```

## Advance

```python
# Heatmap

sns.heatmap(df[['carat','depth','table','price','x','y','z']].corr(), annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmapof correlation Matrix')
plt.show()
```

```python
# Pairplot
sns.pairplot(data=df,hue='cut')
plt.title('Pair Plot of diamond attributes')
plt.show()
```

```python
g = sns.FacetGrid(data=df, col='color', col_wrap=3)
g.map(plt.hist,'price',bins=10,edgecolor='black')
plt.show()
```

## Enhancing Plots with seaborn for Ready-Report Visualization.

### Settings Seaborn Styles and contexs

```python
# Setting Seaborn Style
sns.set_style('whitegrid') # Options for Style : white, dark, whitegrid, darkgrid, ticks :-> According this it will change background grid.
sns.set_context('notebook') # Option for Context : paper,notebook,talk,poster   :-> According this style it will change visualization of plot. 

sns.scatterplot(data=df,x='carat',y='price')
plt.title('Scatter Plot with whitegrid Style and Notebook Context')
plt.show()
```

```python
# Using Dark background style
sns.set_style('dark')

sns.lineplot(data=df,x='carat',y='price')
plt.title('Line Plot With Dark Background')
plt.show()
```

```python
sns.set_context('talk',font_scale=0.5)

sns.boxplot(data=df,x='cut',y='price')
plt.title('Box Plot with Talk Content and large Font Scale')
plt.show()
```

```python
# Custom Color Palette

custom_colour = sns.color_palette(['#FF5733','#33FFCE','#335BFF','#FF33A1','#9B33FF'])
sns.set_palette(custom_colour)
sns.countplot(data=df,x='cut',hue='color')
plt.title('Count Plot with Custom Palette')
plt.show()
```
