```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Bacis

### Task 1: Load Dataset
    Data: tips.csv
    
    Goal: Load and display the first 5 rows using pandas.

```python
df = pd.read_csv('Practice_Datasets/tips_sample.csv')
print(df.head())
```

### Task 2: Scatter Plot
    Data: tips.csv
    
    Goal: Plot total_bill vs tip using sns.scatterplot().

```python
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Task 3: Histogram
    Data: tips.csv
    
    Goal: Plot histogram of total_bill.

```python
plt.figure(figsize=(12,8))

sns.histplot(data=df, x='total_bill',bins=20,edgecolor='black')
plt.title('Histogram Of Total Bill')
plt.show()
```

### Task 4: Box Plot
    Data: tips.csv
    
    Goal: Plot a boxplot of tip.

```python
sns.boxplot(data=df, x='tip')
plt.title('Box Plot Of Tip')
plt.xlabel('Tip')
plt.show()
```

### Task 5: Count Plot
    Data: tips.csv
    
    Goal: Show count of records for each day.

```python
sns.countplot(data=df,x='day')
plt.title('Count Plot of Count Records for Each Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

### Task 6: Bar Plot
    Data: tips.csv
    
    Goal: Compare average tip by sex.

```python
sns.barplot(data=df, x='sex', y='tip', estimator='mean', ci='sd')  # 'ci' adds error bars (standard deviation)
plt.title('Average Tip by Sex')
plt.xlabel('Sex')
plt.ylabel('Average Tip')
plt.show()
```

## Intermediate

### Task 7: Violin Plot
    
    Data: tips.csv
    
    Goal: Create a violin plot for tip vs day.

```python
sns.violinplot(data=df,x='tip',y='day')
plt.title('Violin Plot of Tip and Day')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

### Task 8: Pair Plot
    
    Data: tips.csv
    
    Goal: Plot pairwise relationships using sns.pairplot().

```python
sns.pairplot(data=df,hue='sex')
plt.suptitle('Pair Plot of Tips Dataset',y=1.02)  # Adjust y for better placement of title
plt.show()
```

### Task 9: Hue Parameter in Scatter Plot
    
    Data: tips.csv
    
    Goal: Plot total_bill vs tip with hue based on smoker.

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='smoker')
plt.title('Scatter Plot of Total bill and Tip with Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Task 10: Style and Palette
    
    Data: tips.csv
    
    Goal: Customize scatter plot using style and palette.

```python
palette = sns.palettes.SEABORN_PALETTES
print(palette)
```

```python
sns.set_style('whitegrid')

custom_color = sns.color_palette(['#4878D0', '#6ACC64', '#D65F5F', '#956CB4'])

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='day', y='tip', hue='day', palette=custom_color, s=100, edgecolor='black')  # s=100 is for size of node.
plt.title('Scatter Plot of Tip vs Day with Custom Style and Palette')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount')
plt.legend(title='Day')
plt.tight_layout()
plt.show()
```

### Task 11: Swarm Plot
    
    Data: tips.csv
    
    Goal: Plot a swarm plot of day vs total_bill.

```python
sns.swarmplot(data=df,x='day',y='total_bill')
plt.title('Swarm Plot')
plt.show()
```

### Task 12: Correlation Heatmap
    
    Data: tips.csv
    
    Goal: Plot heatmap of correlation matrix.

```python
sns.heatmap(df[['total_bill', 'tip','size']].corr(),cmap='coolwarm',annot=True, linewidths=0.5)
plt.title('Heatmap of Numerical Features') 
plt.show()
```

### Task 13: Linear Regression Plot (lmplot)
    
    Data: tips.csv
    
    Goal: Show linear regression between total_bill and tip.

#### Linear regression is a method to model the relationship between two variables by fitting a straight line through the data points.

#### It helps predict the value of one variable (like a tip) based on another (like total bill).

```python
sns.lmplot(data=df, x='total_bill', y='tip', height=6, aspect=1.2)
plt.title('Linear Regression: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

# Advance

### Task 14: Combined Box + Swarm Plot
    
    Data: tips.csv
    
    Goal: Combine boxplot and swarmplot for day vs tip.

```python
sns.boxplot(data=df, x="day", y="tip", width=0.5, palette="pastel", hue='day',legend=False)
sns.swarmplot(data=df, x="day", y="tip", color="black", size=3)
plt.title("Boxplot + Swarmplot of Tip by Day")
plt.show()
```

### Task 15: Facet Grid Plot
    
    Data: tips.csv
    
    Goal: Create facet grid plots for total_bill by time and sex.

```python
g = sns.FacetGrid(data=df,row='time',col='sex')
g.map(sns.lineplot, 'day', 'total_bill') 
g.set_axis_labels("Table Size", "Total Bill")
g.figure.suptitle("Facet Grid of Total Bill by Time and Sex", fontsize=16)
g.figure.subplots_adjust(top=0.9)  # Adjust space for title
plt.show()
```

### Task 16: Catplot
    
    Data: tips.csv
    
    Goal: Use catplot() to visualize sex vs tip by day.

```python
sns.catplot(data=df, x='sex', y='tip', hue='day', kind='box')

plt.title("Tip Amount by Sex and Day")
plt.show()
```

### Task 17: Set Themes and Context
    
    Data: tips.csv
    
    Goal: Apply themes using sns.set_style() and sns.set_context().

```python
sns.set_style('whitegrid')
sns.set_context('talk')

sns.boxplot(x='day', y='total_bill', data=df, palette='pastel',hue='day')
plt.title("Total Bill by Day")

plt.show()
```

### Task 18: Multi-Plot Grid
    
    Data: tips.csv
    
    Goal: Use sns.FacetGrid() to map multiple plots manually.

```python
g = sns.FacetGrid(df, col="sex", row="time", margin_titles=True,height=4, aspect=1)
g.map(sns.lineplot, "total_bill", "tip", color="blue")
g.map(sns.histplot, "total_bill", kde=True, color="green")
g.figure.suptitle("Total Bill vs Tip by Sex and Time", fontsize=16)
g.figure.subplots_adjust(top=0.86)

plt.show()
```

### Task 19: Save Plot as Image
    
    Data: tips.csv
    
    Goal: Save your Seaborn plot using plt.savefig().

```python
g = sns.FacetGrid(df, col="sex", row="time", margin_titles=True,height=4, aspect=1)
g.map(sns.lineplot, "total_bill", "tip", color="blue")
g.map(sns.histplot, "total_bill", kde=True, color="green")
g.figure.suptitle("Total Bill vs Tip by Sex and Time", fontsize=16)
g.figure.subplots_adjust(top=0.86)

plt.savefig('FacetGrid.png')
```

### Task 20: Subplot Layout
    
    Data: tips.csv
    
    Goal: Arrange multiple Seaborn plots using plt.subplots().

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

axes[0].scatter(df['total_bill'],df['tip'])
axes[0].set_title('Scatter Plot of Total Bill and Tip', fontsize=14)
axes[0].set_xlabel('Total Bill', fontsize=12)
axes[0].set_ylabel('Tip', fontsize=12)

axes[1].hist(df['day'],bins=10,edgecolor='black')
axes[1].set_title('Histogram of Days', fontsize=14)
axes[1].set_xlabel('Sex', fontsize=12)
axes[1].set_ylabel('Tip', fontsize=12)

plt.show()
```
