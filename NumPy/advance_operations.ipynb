## Advance Operations with Business Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [resturant_id,2021,2022,2023,2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],    # Parasdise Biryani
    [2, 120000, 140000, 160000, 190000],    # Beijing Bites
    [3, 200000, 230000, 260000, 300000],    # Pizza Hub
    [4, 180000, 210000, 240000, 270000],    # Burger Point
    [5, 160000, 185000, 205000, 230000],    # Chai Point
])

print("=== Zomato sales analysis ===")
print()
print("Shape of an array is",sales_data.shape)
print("\n Sample data for first 3 resturant :\n",sales_data[:3])
```

```python
# axis = 0 work vertically(Column)
# axis = 1 work Horizontally(Row)

#total sales per year using row
yearly_total_row = np.sum(sales_data[:,1:],axis=1)  
print("Yearly total by row is ",yearly_total_row)


#total sales per year using column
yearly_total_column = np.sum(sales_data[:,1:],axis=0)
print("Yearly total by column is ",yearly_total_column)


# Total sales for all year for all resturants
print("Total sales for all year for all resturants",np.sum(yearly_total_column))



# Here,in sales data we  want all rows but don't want all column so therefore we write
# sales_data[:,1:]. here first ":" represent row and second "1:" represent column
# we dont want first column that's  why we use slicing and remove first column
```

```python
# Minimum sale
min_sale = np.min(sales_data[:,1:],axis=1)
print("Minimum sale is",min_sale)

# Maximum sale per resturant.
max_sale = np.max(sales_data[:,1:],axis=1)
print("Maximum sales are ",max_sale)

#Maximum sales per year.
max_sale_year = np.max(sales_data[:,1:],axis=0)
print("Maximum sales per year is", max_sale_year)

# Average sales per resturant.
average_sales = np.average(sales_data[:,1:],axis=0)
print("Average sales per resturant is ",average_sales)
```

```python
# Cumulative sums : display the total sum of data as it grows with time

cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

# Create visual representation 
plt.figure(figsize=(6,6))
plt.plot(np.mean(cumsum, axis=0)) # axis=0 means it will find mean year 
plt.title("Average cumulative sales accross all resturant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
