```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [resturant_id,2021,2022,2023,2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],    # Parasdise Biryani
    [2, 120000, 140000, 160000, 190000],    # Beijing Bites
    [3, 200000, 230000, 260000, 300000],    # Pizza Hub
    [4, 180000, 220000, 240000, 270000],    # Burger Point
    [5, 160000, 185000, 205000, 230000],    # Chai Point
])
```

### 1. Basic Statistics

```python
# What is the total sales of each restaurant across all year?
total_sales = np.sum(sales_data[:,1:],axis=1)
print("Total sales for each resturant is",total_sales)

# What is the average sales per year for each restaurant?
average_sales = np.average(sales_data[:,1:],axis = 1)
print("Average sales for each resturant is",average_sales)
reshape = average_sales.reshape(5,1)
print(reshape)

# What is the maximum and minimum sales for each restaurant?
minimum_sales = np.min(sales_data[:,1:], axis=1)
print("Minimum sales is",minimum_sales)

maximum_sales = np.max(sales_data[:,1:], axis=1)
print("Minimum sales is",maximum_sales)
```

### 2. Indexing & Slicing

```python
# Extract sales data for the second year.

print(sales_data[:,2:3])

# Extract sales data for "Pizza Hub" only.
print("Data of Pizza Hub",sales_data[2,1:])

# Retrieve the sales data for "Burger Point" in the third year.
print("Data of Burger Point in the third quarter",sales_data[3,3:4])
```

### 3. Summation & Aggregation

   

```python
# What is the total revenue generated across all restaurants in all year?
resturant_total_sales = np.sum(sales_data[:,1:],axis=1)

# print(resturant_total_sales)
reshape = resturant_total_sales.reshape(5,1)
print(reshape)

print()

# Find the total sales for each year.
total_sales_year = np.sum(sales_data[:,1:],axis=0)
print("Total sale per year is\n",total_sales_year)
```


### 4. Sorting & Ranking


```python
# Rank the restaurants based on total sales.
based_total_sale = np.sum(sales_data[:,1:],axis=1)
print("Total sales",based_total_sale)

sorted = np.sort(based_total_sale)
print("Sorted sales",sorted)

print()

# Sort the restaurants based on their highest sales in any year.
higest_sales_year = np.sum(sales_data[:,1:],axis=0)
print("Higest Sales are",higest_sales_year)
```


### 5. Comparison & Filtering


```python
# Find the restaurants where sales in the fourth year exceeded 250000.
finding= sales_data[:,4]
print(finding)

find_resturant = np.where(finding > 250000)
print(finding[find_resturant])



# Identify the restaurants that had sales below 200000 in any year.

sales = sales_data[:,1:]

sales_below = np.where(sales_data < 200000)
print(sales_data[sales_below])


sales_below = np.any(sales < 200000, axis=1)
print(sales_data[sales_below,0])
```
