```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.random.random((3,3))
array3 = np.zeros((4,4))
```

### Save array 

```python
np.save('array1.npy',array1)
np.save('array2.npy',array2)
np.save('array3.npy',array3)
```

### Load and display value which is stored in file 

```python
loaded_array1 = np.load('array1.npy')
print(loaded_array1)
```

### Load Image an save into array format

```python
image = Image.open("numpy-logo.png")
image_array = np.array(image)

np.save("numpy-logo.npy",image_array)
```

```python
try:
    logo = np.load('numpy-logo.npy')

    #Display Logo
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.title("NumPy Logo")
    plt.grid(False)
    plt.imshow(logo)
    

except FileNotFoundError:
    print("Numpy Logo file does not found")
```
